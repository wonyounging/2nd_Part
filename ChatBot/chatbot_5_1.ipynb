```python
class Text_Similar:
    def word_ngram(bow, num_gram):
        text = tuple(bow)
        ngrams = [text[x:x + num_gram] for x in range(0,len(text))]
        return(ngrams)

    def similarity(doc1, doc2):
        cnt = 0
        for token in doc1:
            if token in doc2:
                cnt = cnt + 1
        return cnt/len(doc1)

class Input_Sentence:
    def in_sentence():
        return input("3개 이상의 단어로 구성된 문장을(를) 입력하세요 : ")
    
    def in_sentence_filter(sentence):
        if len(sentence) > 0:
            komoran = Komoran()
            bow = komoran.nouns(sentence)
            if len(bow) < 3 and len(bow) > 0:
                print('3개 이상의 단어로 구성해주세요.')
            else:
                return True
        else:
            print('공백은 입력할 수 없습니다.')

if __name__ == "__main__":
    
    while(1):
        ips = Input_Sentence
        sentence1 = ips.in_sentence()
        if ips.in_sentence_filter(sentence1) == True:
            bow1 = komoran.nouns(sentence1)
            break

    while(1):
        ips = Input_Sentence
        sentence2 = ips.in_sentence()
        if ips.in_sentence_filter(sentence2) == True:
            bow2 = komoran.nouns(sentence2)
            break

    ts = Text_Similar
    doc1 = ts.word_ngram(bow1, 2)
    doc2 = ts.word_ngram(bow2, 2)

    print(doc1)
    print(doc2)

    r1 = ts.similarity(doc1, doc2)

    print(f'문장 1과 문장 2의 유사도 : {r1}')
```

# 코사인 유사도 계산

```python
from konlpy.tag import Komoran
import numpy as np
from numpy import dot
from numpy.linalg import norm
```

```python
# 코사인 유사도 계산_1
def cos_sim(vec1, vec2):
    return dot(vec1, vec2) / (norm(vec1) * norm(vec2))

# TDM 만들기_2
def make_term_doc_mat(sentence_bow, word_dics):
    freq_mat = {}
    for word in word_dics:
        freq_mat[word] = 0
    for word in word_dics:
        if word in sentence_bow:
            freq_mat[word] += 1
    return freq_mat

# 단어 벡터 만들기_3
def make_vector(tdm):
    vec = []
    for key in tdm:
        vec.append(tdm[key])
    return vec

# 문장 정의
sentence1 = "6월에 뉴턴은 선생님의 제안으로 트리니티에 입학하였다."
sentence2 = "6월에 뉴턴은 선생님의 제안으로 대학교에 입학하였다."
sentence3 = "나는 맛있는 밥을 뉴턴 선생님과 함께 먹었습니다."

# 형태소분석기를  이용해 단어 묶음 리스트 생성_4
komoran = Komoran()
bow1 = komoran.nouns(sentence1)
bow2 = komoran.nouns(sentence2)
bow3 = komoran.nouns(sentence3)

# 단어 묶음 리스트르 하나로 합침_5
bow = bow1 + bow2 + bow3

# 단어 묶음에서 중복 제거하여 단어사전 구축_6
word_dics = []
for token in bow:
    if token not in word_dics:
        word_dics.append(token)

# 문장 별 단어 문서 행렬 계산_7
freq_list1 = make_term_doc_mat(bow1, word_dics)
freq_list2 = make_term_doc_mat(bow2, word_dics)
freq_list3 = make_term_doc_mat(bow3, word_dics)
print(freq_list1)
print(freq_list2)
print(freq_list3)

# 문장 벡터 생성_8
doc1 = np.array(make_vector(freq_list1))
doc2 = np.array(make_vector(freq_list2))
doc3 = np.array(make_vector(freq_list3))

# 코사인 유사도 계산_9
r1 = cos_sim(doc1, doc2)
r2 = cos_sim(doc3, doc1)

print("{}%".format(round(r1*100,2)))
print("{}%".format(round(r2*100,2)))
```

```python
class Text_Similar:
    def cos_sim(vec1, vec2):
        return dot(vec1, vec2) / (norm(vec1) * norm(vec2))

    def make_term_doc_mat(sentence_bow, word_dics):
        freq_mat = {}
        for word in word_dics:
            freq_mat[word] = 0
        for word in word_dics:
            if word in sentence_bow:
                freq_mat[word] += 1
        return freq_mat

    def make_vector(tdm):
        vec = []
        for key in tdm:
            vec.append(tdm[key])
        return vec
    
    def make_dics(bow):
        word_dics = []
        for token in bow:
            if token not in word_dics:
                word_dics.append(token)
        return word_dics

class Input_Sentence:
    def in_sentence():
        return input("3개 이상의 단어로 구성된 문장을(를) 입력하세요 : ")
    
    def in_sentence_filter(sentence):
        if len(sentence) > 0:
            komoran = Komoran()
            bow = komoran.nouns(sentence)
            if len(bow) < 3 and len(bow) > 0:
                print('3개 이상의 단어로 구성해주세요.')
            else:
                return True
        else:
            print('공백은 입력할 수 없습니다.')

if __name__ == "__main__":
    komoran = Komoran()
    while(1):
        ips = Input_Sentence
        sentence1 = ips.in_sentence()
        if ips.in_sentence_filter(sentence1) == True:
            bow1 = komoran.nouns(sentence1)
            break

    while(1):
        ips = Input_Sentence
        sentence2 = ips.in_sentence()
        if ips.in_sentence_filter(sentence2) == True:
            bow2 = komoran.nouns(sentence2)
            break
    
    bow = bow1 + bow2

    ts = Text_Similar
    word_dics = ts.make_dics(bow)

    freq_list1 = ts.make_term_doc_mat(bow1, word_dics)
    freq_list2 = ts.make_term_doc_mat(bow2, word_dics)
    print(freq_list1)
    print(freq_list2)

    doc1 = np.array(ts.make_vector(freq_list1))
    doc2 = np.array(ts.make_vector(freq_list2))
    print(doc1)
    print(doc2)

    r1 = ts.cos_sim(doc1, doc2)

    print(f'문장 1과 문장 2의 유사도 : {round(r1*100,2)}%')
```
