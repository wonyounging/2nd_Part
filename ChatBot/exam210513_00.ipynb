```python
import datetime
from datetime import datetime, date
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
import re

pd.set_option('display.notebook_repr_html', True)
pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 10)
pd.set_option('display.width', 250)
%matplotlib inline 

# 요구사항 1-1 ~ 1-2
f = open('./data/filetest.txt', 'r', encoding='utf8')
txt_t1 = f.read()
txt_t1 = txt_t1.split('\n')
ls1 = txt_t1[1:-1]
# ls1

# 요구사항 1-3
re_ls1 = []
for i in ls1:
    re_ls1.append(re.sub(r"[a-zA-Z]", "", i))
re_ls1

# 요구사항 1-4
ls2 = []
for i in re_ls1:
    ls2.append(i.split('-'))
ls2

# 요구사항 1-5
name = []
number = []
for i in ls2:
    name.append(i[0])
    number.append(i[1])
print(name)
print(number)

# 요구사항 1-6 실행결과
df = pd.DataFrame({'이름':name, '학번':number})
df = df.set_index('학번')
df

# 요구사항 최종 실행결과
print(df.iloc[::2])
print('-'*20)
print(df.iloc[1::2])
```
