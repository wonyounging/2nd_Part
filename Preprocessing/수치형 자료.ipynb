```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/titanic.csv')
data.head(5)
```

```python
def normal(data):
    fare_nor_data = []
    fare_max = data.Fare.max()
    fare_min = data.Fare.min()
    for i in range(len(data)):
        fare_nor_data.append(round((data.iloc[i].Fare- fare_min) / (fare_max -fare_min),6))
    return fare_nor_data
    
def std(data):
    fare_std_data = []
    fare_avg = np.mean(data.Fare)
    fare_std = np.std(data.Fare)
    for i in range(len(data)):
        fare_std_data.append(round(((data.iloc[i].Fare - fare_avg) / fare_std),6))
    return fare_std_data

data['Fare_nor'] = normal(data)
data['Fare_std'] = std(data)
print("변환 전 : \n{}\n".format(data.Fare.head(5)))
print("정규화변환 후 : \n{}\n".format(data.Fare_nor.head(5)))
print("표준화변환 후 : \n{}\n".format(data.Fare_std.head(5)))
```
